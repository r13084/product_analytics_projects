## Предобработка данных

### Обработка дубликатов в первичных данных

Найдём и посчитаем полные дубликаты в датафрейме `clients_solvency`, то есть такие пары строк, где значения по всем имеющимся признакам совпадают, а также выведем их долю от общего количества строк в таблице, однако перед тем, как это сделать, необходимо все значения в столбцах с типом данных *object* привести к нижнему регистру, чтобы избавиться от дублирующих названий, записанных с использованием заглавных букв.

```python
#Напишем цикл, где сначала с помощью метода unique() посчитаем уникальные названия каждого нужного столбца, 
#затем воспользуемся методом str.lower() и снова посчитаем уникальные названия 
for column in ['education', 'family_status', 'gender', 'income_type', 'purpose']:
    print('Уникальных значений в столбце "{}" до:{:.0f}'.format(column, len(clients_solvency[column].unique())))
    clients_solvency[column] = clients_solvency[column].str.lower()
    print('Уникальных значений в столбце "{}" после:{:.0f}'.format(column, len(clients_solvency[column].unique())))
```

```python
#Посчитаем кол-во полных дубликатов, используя поочерёдно методы duplicated() и sum()
print('Кол-во полных дубликатов: {:.0f}'.format(clients_solvency.duplicated().sum()),
     'Доля полных дубликатов от общего кол-ва строк: {:.2%}'.format(
     clients_solvency.duplicated().sum() / clients_solvency.shape[0]), 
     sep='\n')
```

Так как дубликаты ухудшают репрезентативность данных, могут искажать форму их распределения и вызвать смещение статистических оценок, то в общем случае их принято удалять. В нашей ситуации у объектов наблюдения нет уникальных идентификаторов, то есть нет, например, номера и серии паспорта или хотя бы полного имени, поэтому можно считать, что это всё уникальные клиенты, однако, так как признаков (столбцов), по которым осуществляется наблюдение за объектами, достаточно много, то сложно поверить, что это уникальные заявители. Скорее всего причиной появления таких дубликатов стала либо техническая ошибка в базе данных, либо человеческая невнимательность. В связи с этим считаем, что данные дубликаты целесообразно удалить, хотя даже если это всё уникальные люди, то это удаление не повлияет на выводы исследования, так как доля этих строк совсем незначительная.

```python
#С помощью метода drop_duplicates() удалим дубликаты, а затем избавимся от индексов удалённых строк
#и сделаем нумерацию заново методом reset_index()
print('Кол-во наблюдений в таблице "clients_solvency" до удаления полных дубликатов: {}'.format(clients_solvency.shape[0]))
clients_solvency = clients_solvency.drop_duplicates().reset_index(drop=True)
print('Кол-во наблюдений в таблице "clients_solvency" после удаления полных дубликатов: {}'.format(clients_solvency.shape[0]))
```

### Поиск аномалий в данных и их обработка

На этом этапе попытаемся найти и оценить количество значений для каждого признака датафрейма `clients_solvency`, которые либо сильно отличаются от общей массы, либо вообще не характерны для объектов исследования. Анализ будем проводить отдельно по каждому столбцу, кроме столбцов *education_id*, *family_status_id*, *purpose*, если данных достаточно, то будем их обрабатывать, если нет, то сделаем это в следующих пунктах.

Напишем функцию *counts_and_percent()*, которая будет выводить для конкретного столбца таблицы `clients_solvency` датафрейм, в котором перечислены уникальные значения, сколько раз они встречаются и их доля от общего количества записей.

```python
def counts_and_percent(certain_column):
    return pd.DataFrame({'counts': certain_column.value_counts(),
                        'percent': certain_column
                         .value_counts(normalize=True)
                         .mul(100).round(2).astype(str) + '%'})
```

**Столбец "children"**

```python
#Применим написанную нами функцию counts_and_percent() к столбцу children
counts_and_percent(clients_solvency['children'])
```

Из необычного видим, что у 76 клиентов по 20 детей, а у 47 клиентов "-1". 20 детей, как уже отмечали, это очень редко, а "-1" это невозможно, поэтому предполагаем, что это просто опечатки, сделанные при подаче заявления. Заменим их на "2" и "1" соответственно (даже если там на самом деле другие значения, то наша замена не повлияет в дальнейшем на общие характеристики распределения данных, так как аномальных значений в сумме тут меньше одного процента).

```python
clients_solvency.loc[clients_solvency['children'] == 20, 'children'] = 2
clients_solvency.loc[clients_solvency['children'] == -1, 'children'] = 1

#Проверим, как прошла замена
counts_and_percent(clients_solvency['children'])
```

### Обработка пропусков

Если на этапе знакомства с входными данными для нашего проекта мы отметили, что пропущенные значения присутствуют в двух столбцах таблицы `clients_solvency`, то на этапе поиска и обработки аномалий в данных, а именно в пункте про столбцы *total_income* и *days_employed*, мы приняли решение, что имеет смысл обрабатывать пропуски только в столбце *total_income*.

Нам кажется, что с точки зрения простоты и степени влияния на основные параметры распределения рассматриваемой переменной наиболее подходящим способом заполнения пропусков является использование либо среднего, либо медианы. Однако среднее может быть не совсем объективным отражением данных, так как на него могут влиять выбросы, которые, как мы уже определили, присутствуют в достаточном количестве, поэтому будем использовать медиану.

Для лучшего соответствия будем заполнять пропуски медианой, используя группировку по двум признакам таблицы `clients_solvency`. Двух признаков, считаем, будет достаточно, так как, если использовать больше, то группы получатся небольшими, и тогда доверять их характеристикам будет сложно.

С нашей точки зрения, наиболее подходящими будут признаки *income_type* и *dob_years*, так как остальные параметры либо в меньшей степени связаны с уровнем дохода, либо могут в достаточном количестве случаев принимать одни и те же значения при разном заработке. Например, существенно помогать содержать детей (столбец *children*) могут родители родителей. Семейное положение (столбец *family_status*) - это, нам кажется, во многои параллельный профессиональной деятельности процесс. 

Возраст клиента банка (столбец *dob_years*) в некоторой степени заменит  информацию из забракованного нами признака *days_employed*. В совокупности с возрастом заявителя данные столбца *income_type* могут дать нам представление об уровне дохода в рамках каждого типа занятости.

### Лемматизация

На этапе знакомства с входными данными мы отметили, что клиенты банка при оформлении заявлений по-разному описывали по сути одну и ту же цель (колонка *purpose*), для реализации которой им требовался кредит. На этапе обработки аномалий после приведения всех значений к нижнему регистру мы определили, что количество уникальных значений для признака *purpose* равно 38. Этот факт затрудняет нам верно категоризировать клиентов, чтобы дать ответ на один из вопросов исследования, так как по содержанию одной причине могут соответствовать сразу несколько формулировок. Чтобы избежать такой ситуации, мы поступим следующим образом:

* просмотрим список уникальных формулировок, почему нужен кредит, и выделим уникальные цели;
* к списку уникальных значений признака *purpose* применим такой алгоритм для поиска слов, записанных в разных формах, как лемматизация, и приведём все слова к их словарной форме (лемме);
* подсчитаем частоту появления лемматизированных слов, затем соотнесём выделенные нами уникальные цели с наиболее частыми леммами;
* проведём лемматизацию для каждой записи признака *purpose*, по ключевым словам определим уникальную цель получения клиентом кредита.

```python
#Перечислим уникальные значения признака purpose
for purpose in clients_solvency['purpose'].unique():
    print(purpose)
```

Из всех формулировок, почему клиенту требуется кредит, мы можем выделить вне зависимости от формы четыре общие уникальные цели:

1) Операция с недвижимостью;  
2) Проведение свадьбы;    
3) Сделка с автомобилем;   
4) Получение образования.   

Каждая из выделенных нами целей объединяет в себе более специфические причины, то есть, например под "операцией с недвижимостью" может пониматься ремонт, строительство, покупка собственного жилья или покупка коммерческой недвижимости для сдачи и другое, связанное с недвижимостью. Мы считаем, что за счёт таких агрегированных целей доверие к будущим результатам исследования будет высоким, так как каждой такой причине соответствует достаточное количество наблюдений. 

```python
#Лемматизируем уникальные значения признака purpose методами из библиотеки pymystem3
lemmas = Mystem().lemmatize(' '.join(clients_solvency['purpose'].unique()))

#Подсчитываем частоту появления лемм с помощью словаря модуля Collections и сортируем его методом most_common()
Counter(lemmas).most_common()
```

Сопоставим выделенные нами уникальные цели получения кредита с наиболее частыми леммами.  

1) Операция с недвижимостью: "*недвижимость*", "*жилье*";  
2) Проведение свадьбы: "*свадьба*";  
3) Сделка с автомобилем: "*автомобиль*";  
4) Получение образования: "*образование*".  

```python
#Пролемматизируем каждое значение столбца purpose, результат сохраним в новой колонке purpose_lemmas
clients_solvency['purpose_lemmas'] = clients_solvency['purpose'].apply(Mystem().lemmatize)

#Проверим результат
clients_solvency[['purpose', 'purpose_lemmas']].head(7)
```

```python
#Напишем функцию, которая будет среди лемм находить ключевые слова и возвращать уникальную цель для получения кредита
def purpose_grouping(list_of_lemmas):
    if ('недвижимость' in list_of_lemmas) or ('жилье' in list_of_lemmas):
        return 'операция с недвижимостью'
    if 'свадьба' in list_of_lemmas:
        return 'проведение свадьбы'
    if 'автомобиль' in list_of_lemmas:
        return 'сделка с автомобилем'
    if 'образование' in list_of_lemmas:
        return 'получение образования'
```

```python
#Применим функцию purpose_grouping к значениям столбца purpose_lemmas, результат сохраним в новой колонке purpose_group
clients_solvency['purpose_group'] = clients_solvency['purpose_lemmas'].apply(purpose_grouping)

#Проверим результат
clients_solvency[['purpose', 'purpose_lemmas', 'purpose_group']].head(7)
```

```python
#Посмотрим теперь, сколько заявок соответствует той или иной цели
display(counts_and_percent(clients_solvency['purpose_group']))

#Проверим состояние датафрейма clients_solvency на данный момент
clients_solvency.info()
```

| Тип семьи     | Кол-во детей в семье |
|:------------- |:---------------------|

**Категоризация признака "total_income"**

В задании к проекту у нас не было уточнений про то, в какой валюте происходит исчисление дохода клиентов банка, гражданами какой страны они являются, в каком регионе проживают, поэтому при делении заявителей на группы по заработку мы не станем использовать или ориентироваться на какие-то внешние статистические отчёты, а будем использовать следующую условную градацию: "низкий", "средний", "выше среднего", "высокий", "очень высокий" доходы.

В категорию "очень высокие" доходы мы отнесём найденные нами на этапе поиска аномалий выбросы, а для остальных четырёх категорий получим промежутки таким образом, чтобы в каждый из них попало примерно равное количество значений выборки за исключением сверхвысоких заработков.

```python
#Выделим промежутки для первых четырёх категорий признака total_income методом qcut(), исключив выбросы с помощью полученной
#нами переменной upper_edge_of_interquartile_range - "верхняя граница межквартального размаха", - и посмотрим на то,
#сколько клиентов соответствуют тому или иному полученному промежутку
pd.qcut(
    clients_solvency
    .loc[clients_solvency['total_income'] <= upper_edge_of_interquartile_range, 'total_income'], q = 4).value_counts()
```

```python
#Сохраним список по порядку перечисленных границ полученных промежутков для категоризации признака total_income
#в переменную intervals
intervals = (pd.qcut(
    clients_solvency
    .loc[clients_solvency['total_income'] <= upper_edge_of_interquartile_range, 'total_income'], q = 4, retbins = True)[1])

#Напишем функцию, которая будет по значению столбца total_income определять, к какой группе по доходу относится клиент банка
def total_income_grouping(amount):
    if amount >= intervals[0] and amount <= intervals[1]:
        return 'низкий доход'
    if amount > intervals[1] and amount <= intervals[2]:
        return 'средний доход'
    if amount > intervals[2] and amount <= intervals[3]:
        return 'выше среднего доход'
    if amount > intervals[3] and amount <= intervals[4]:
        return 'высокий доход'
    if amount > intervals[4]:
        return 'очень высокий доход'
```

```python
#Сохраним результат работы функции total_income_grouping() в новую колонку total_income_group
clients_solvency['total_income_group'] = clients_solvency['total_income'].apply(total_income_grouping)

#Проверим результат
display(clients_solvency[['total_income', 'total_income_group']].head(7))

counts_and_percent(clients_solvency['total_income_group'])
```

Искать ответ на каждый вопрос мы будем с помощью сводных таблиц, в которых по группам отдельно взятого признака датафрейма `clients_solvency` найдём соотношение между займами, которые были погашены в срок, и займами, по которым у клиентов образовалась задолженность перед банком. Так как выводы, которые мы получим, нужны для построения модели кредитного скоринга - специальной системы, которая оценивает способность потенциального заёмщика вернуть кредит банку, - то долю погашенных в срок займов будем считать вероятностью заёмщика вернуть кредит.

**Есть ли зависимость между количеством детей у клиента и возвратом кредита в срок?**

```python
#Создадим сводную таблицу children_report, где названиями строк становятся значения столбца children_group (по ним же происходит группировка),
#названиями столбцов становятся значения колонки debt (по ним также происходит группировка), 
#к значениям признака total_income (можно было выбрать другой признак) 
#применяется функция подсчёта количества значений count()
children_report = clients_solvency.pivot_table(index = 'children_group', 
                                               columns = 'debt', 
                                               values = 'total_income', 
                                               aggfunc = 'count')

#Изменим названия столбцов: no_debt - нет задолженности, yes_debt - есть задолженность.
children_report.columns = ['no_debt', 'yes_debt']

#Посчитаем долю кредитов, которые вернули в срок
children_report['%_of_paid_loans_on_time'] = children_report['no_debt'] / (children_report['yes_debt'] 
                                                                           + children_report['no_debt'])

#Посчитаем долю кредитов, по которым есть задолженность
children_report['%_of_late_payments'] = children_report['yes_debt'] / (children_report['yes_debt'] 
                                                                       + children_report['no_debt'])

#Настроем формат представления данных таблицы children_report и выведем её
children_report.style.format({'%_of_paid_loans_on_time':'{:.2%}', '%_of_late_payments' : '{:.2%}'})
```

Из полученной нами сводной информации по категориям заявителей с точки зрения количества детей у них не видим какой-то существенной разницы между клиентами без детей, клиентами, у кого один или два ребёнка, и клиентами, имеющими много у себя детей, относительно доли погашенных в срок займов.

### Вывод

Не наблюдаем зависимость между тем, сколько у человека, оформившего кредит в банке, есть детей на иждивении и вероятностью вернуть заимствованные им средства в срок. Другими словами, при прочих равных нет разницы, кому стоит выдать кредит: гражданину без детей, гражданину с одним ребёнком или двумя или гражданину, у которого три и более ребёнка находятся на содержании.

**Есть ли зависимость между семейным положением и возвратом кредита в срок?**

```python
#Создадим сводную таблицу family_status_report
family_status_report = clients_solvency.pivot_table(index = 'family_status', 
                                                    columns = 'debt', 
                                                    values = 'total_income', 
                                                    aggfunc = 'count')

family_status_report.columns = ['no_debt', 'yes_debt']

family_status_report['%_of_paid_loans_on_time'] = family_status_report['no_debt'] / (family_status_report['yes_debt'] 
                                                                                     + family_status_report['no_debt'])

family_status_report['%_of_late_payments'] = family_status_report['yes_debt'] / (family_status_report['yes_debt'] 
                                                                                 + family_status_report['no_debt'])

family_status_report.style.format({'%_of_paid_loans_on_time':'{:.2%}', '%_of_late_payments' : '{:.2%}'})
```
